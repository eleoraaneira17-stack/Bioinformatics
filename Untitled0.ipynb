```python
def cek_nutrisi(kadar_n=None, kondisi_daun=None):
    status = "Tidak diketahui"
    rentang_kadar_n = "Tidak diketahui"

    if kadar_n is not None:
        if 15000 <= kadar_n <= 20000:
            kondisi_daun = "kuning"
        elif 20000 <= kadar_n <= 40000:
            kondisi_daun = "hijau"
        elif 40000 <= kadar_n <= 60000:
            kondisi_daun = "hijau gelap"
        else:
            kondisi_daun = "Tidak Realistis atau Toxic"

        if kondisi_daun == "kuning":
            status = "kekurangan nitrogen"
        elif kondisi_daun == "hijau":
            status = "Nitrogen cukup"
        elif kondisi_daun == "hijau gelap":
            status = "kelebihan nitrogen"
        else:
            status = "Tidak Realistis atau Toxic"

    elif kondisi_daun is not None:
        kondisi_daun = kondisi_daun.lower().strip()
        if kondisi_daun == "kuning":
            status = "kekurangan nitrogen"
            rentang_kadar_n = "15000 - 20000 mg/kg"
        elif kondisi_daun == "hijau":
            status = "Nitrogen cukup"
            rentang_kadar_n = "20000 - 40000 mg/kg"
        elif kondisi_daun == "hijau gelap":
            status = "kelebihan nitrogen"
            rentang_kadar_n = "40000 - 60000 mg/kg"
        else:
            status = "Tanaman tanpa keterangan"
            rentang_kadar_n = "Tanaman tanpa keterangan"

    return kondisi_daun, status, kadar_n, rentang_kadar_n

print("=== ANALISIS NUTRISI TANAMAN ===")

jumlah = int(input("Masukkan jumlah sampel tanaman: "))

count_kurang = 0
count_cukup = 0
count_lebih = 0
count_Tidak_Realistis_atau_Toxic = 0
count_Tanaman_tanpa_keterangan = 0

for i in range(jumlah):
    print(f"\nData tanaman ke-{i+1}:")
    nama = input("Masukan nama tanaman: ")
    pilihan = input("Pilih input: 1 untuk Kadar Nitrogen, 2 untuk Kondisi Daun: ").strip()

    if pilihan == "1":
        n = float(input("Kadar Nitrogen (mg/kg): "))
        kondisi_daun, status, kadar_n, rentang_kadar_n = cek_nutrisi(kadar_n=n)
        print("nama tanaman:", nama)
        print("Kadar Nitrogen:", kadar_n, "mg/kg")
        print("Kondisi daun:", kondisi_daun)
        print("Status nutrisi:", status)

    elif pilihan == "2":
        daun = input("Deskripsikan kondisi daun apakah berwarna kuning / hijau / hijau gelap: ").strip()
        kondisi_daun, status, kadar_n, rentang_kadar_n = cek_nutrisi(kondisi_daun=daun)
        print("nama tanaman:", nama)
        print("Kondisi daun:", kondisi_daun)
        print("Rentang Kadar Nitrogen:", rentang_kadar_n)
        print("Status nutrisi:", status)

    else:
        print("Pilihan tidak valid. Lewati sampel ini.")
        continue

    if status == "kekurangan nitrogen":
        count_kurang += 1
    elif status == "Nitrogen cukup":
        count_cukup += 1
    elif status == "kelebihan nitrogen":
        count_lebih += 1
    elif status == "Tidak Realistis atau Toxic":
        count_Tidak_Realistis_atau_Toxic += 1
    elif status == "Tanaman tanpa keterangan":
        count_Tanaman_tanpa_keterangan  += 1

print("\n=== RANGKUMAN AKHIR ===")
print("Tanaman kekurangan N                 :", count_kurang)
print("Tanaman cukup N                      :", count_cukup)
print("Tanaman kelebihan N                  :", count_lebih)
print("Tanaman tidak realistis atau toxic   :", count_Tidak_Realistis_atau_Toxic)
print("Tanaman tanpa keterangan             :", count_Tanaman_tanpa_keterangan)

print("\nAnalisis selesai!")
```
